```python
# Correlations
import matplotlib.pyplot as plt

def get_corr_map(data):
    corr = data.corr()
    plt.figure(figsize=(8, 8), dpi=100, facecolor='w', edgecolor='k')
    colormap = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(colormap)
    plt.title('Correlation Map')
    plt.show()
    return corr
```

```python
# In our previous example using flattern data, we found that there is no strong correlation between bool resources and other numerical values. We conclude that bool resources are randomly scattered across the map
corr = get_corr_map(data)
```

```python
corr
```

```python
print('Correlation Orders')
for attr in corr:
    print(corr[attr].sort_values(ascending=False))
    print()
```

# Optimize the dataset

```python
# Drop unuseful columns for training
drop_cols = ['Map Name', 'Named Location', 'Rare Metals']
data_dropped = data.drop(drop_cols, axis=1)
data_dropped = data_dropped.loc[:, :'Cold Waves']
data_dropped
```

```python
# Categorical features
s = (data_dropped.dtypes == 'object')
object_cols = list(s[s].index)
print("Categorical attributes")
print(object_cols)
```

```python
# One-hot encoding categorical data
data_encoded = data_dropped.copy()
for object_col in object_cols:
    col = pd.get_dummies(data_dropped[object_col], prefix=object_col)
    data_encoded = pd.concat([data_encoded, col], axis=1)
data_encoded = data_encoded.drop(object_cols, axis=1)
data_encoded.iloc[:5, -8:]
```

```python
# Parse chanllge value
# 140, 100, 220, 180, 240
data_parsed = data_encoded.copy()
data_parsed.replace(
    {
        'Difficulty Challenge':{
            140:140/1080,
            100:100/1080,
            220:220/1080,
            180:180/1080,
            240:240/1080}},
            inplace=True)
```
